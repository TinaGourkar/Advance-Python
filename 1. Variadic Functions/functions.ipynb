### Create args inside function
- *args - this will act like tuple in python

```python
def f1(*args):   # this tells that we can add multiple inputs
    print(args)
    print(type(args))
    for i in args:
        print(i)
```

```python
f1(7,5)
```

```python
f1('a',5)
```

```python
f1('Tina',9,789654123,'Data analyst')
```

### Given the cart products of a customer perform total billing of products

```python
cart = [
    ("Laptop", 65000,1),
    ("Mouse",1500,2),
    ("Keyboard",2000,1)
]
```

```python
def billing(*args):
    # Initalize total bill to 0
    total_bill = 0

    # Apply for loop on args
    for name, price, qty in args:
        subtotal = price*qty
        total_bill = total_bill + subtotal
        print(f"Name : {name}, Price : {price} INR, Quantity : {qty}")
        print(f"Subtotal : {subtotal} INR")
        print("\n---------------------------------------------------------\n")
    
    # Print and return total bill
    print(f"Total Bill : {total_bill:.2f} INR")
    return total_bill
```

```python
cart
```

```python
t1 = billing(*cart)   # here we provided the created tuple
```

```python
t1
```

```python
t2 = billing(   # we are creating a tuple inside of billing function
       ('Mobile',30000,2),
       ('Screen Guard',1200,3),
       ('Headphones',2000,3),
       ('Ear Buds',5000,2)
)
```

```python
t2
```

### Keyword Arguments
- **kwargs - These act like dictionary in python
- {
            key1 : value1,
            key2 : value2
        }

```python
def f2(**kwargs):     # user defined function to separate the key value in the provided dictionary
    print(kwargs)
    print(type(kwargs))
    for k, v in kwargs.items():
        print(f"Key : {k}, Value : {v}")
```

```python
f2(a=1 ,b=23, name='Tia', job='Data Scientist')
```

### Storing information about different users using **kwargs

```python
def create_user(**kwargs):
    user ={
        "u_name" : None,
        "age" : None,
        "email" : None
    }
    user.update(kwargs)
    return user
```

```python
u1 = create_user(    
    u_name = "Lia",
    age = 29
)
```

```python
u1 # returning values according to the parameters provided in the dictionary
```

```python
u2 = create_user(    # here we provided the given parameters
    u_name = 'Mary',
    age = 25,
    email = 'mary@gmail.com'
)
```

```python
u2
```

```python
u3 = create_user(     # here we provided more parameters in our dictionary which is possible 
    u_name = 'James',
    age = 24,
    email = 'james@gmail.com',
    job = 'HR',
    contact = 35497852
)
```

```python
u3
```
